#### Creating new relevant columns

```python
data
```

```python
#if the breed ends with " / MIX" then remove the " / MIX" and in a new column called "Mixed" put true
data['Mixed'] = data['breed'].str.endswith(' / MIX')
data['breed'] = data['breed'].str.replace(' / MIX', '')
data
```

```python
#create a new column called time_in that is indate - outdate
data['days_in'] = pd.to_datetime(data['outdate']) - pd.to_datetime(data['indate'])
data['days_in'] = data['days_in'].dt.days
data
```

```python
#create two new columns in_season and out_season that are fall, winter, spring, summer based on the month of indate and outdate
data['in_season'] = pd.to_datetime(data['indate']).dt.month.map({1: 'WINTER', 2: 'WINTER', 3: 'SPRING', 4: 'SPRING', 5: 'SPRING', 6: 'SUMMER', 7: 'SUMMER', 8: 'SUMMER', 9: 'FALL', 10: 'FALL', 11: 'FALL', 12: 'WINTER'})
data['out_season'] = pd.to_datetime(data['outdate']).dt.month.map({1: 'WINTER', 2: 'WINTER', 3: 'SPRING', 4: 'SPRING', 5: 'SPRING', 6: 'SUMMER', 7: 'SUMMER', 8: 'SUMMER', 9: 'FALL', 10: 'FALL', 11: 'FALL', 12: 'WINTER'})
data
```

```python
#drop indate and outdate columns
data.drop(columns=['indate', 'outdate'], inplace=True)
```

```python
#for breeds with multiple breeds, split the breeds into two columns and update mixed column to true if there were false before. If mixed is true, but breed2 is NaN, update to "unknown"
data['breed'] = data['breed'].str.replace(' / ', ',')
data['breed1'] = data['breed'].str.split(',').str[0]
data['breed2'] = data['breed'].str.split(',').str[1]
data['Mixed'] = data['Mixed'] | data['breed'].str.contains(',')
data
```

```python
#if mixed is true, but breed2 is NaN, update to "unknown"
data.loc[(data['Mixed'] == True) & (data['breed2'].isna()), 'breed2'] = 'UNKNOWN'
data
```

```python
#drop breed column
data.drop(columns=['breed'], inplace=True)
#rename breed1 to breed
data.rename(columns={'breed1': 'breed'}, inplace=True)
#renamed breed2 to mixed_breed
data.rename(columns={'breed2': 'mixed_breed'}, inplace=True)
```

```python
#export to csv
data.to_csv('../Data_prep/Louisville_Animal_Cleaned.csv', index=False)
```
