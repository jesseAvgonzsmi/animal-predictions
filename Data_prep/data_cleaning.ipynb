#### Handling nulls

##### Setup

```python
#display count of nulls for each col
data.isnull().sum()
```

```python
data.drop(['jurisdiction','kennel','sourcezipcode','outsubtype','animalid','ObjectId'], axis=1, inplace=True)
```

```python
#drop animals with no outtype
data.dropna(subset=['outtype'], inplace=True)
```

```python
#create a frame that contains all the rows with nulls
null_data = data[data.isnull().any(axis=1)]
null_data.head()
```

```python
data.isnull().sum()
```

##### Null Insubtype

```python
#create a frame that contains all the insubtype nulls
null_insubtype = null_data[null_data['insubtype'].isnull()]
null_insubtype['intype'].value_counts()
```

```python
#fill nulls with 'Unknown'
data['insubtype'].fillna('Unknown', inplace=True)
```

##### Null surreason

```python
null_surreason = null_data[null_data['surreason'].isnull()]
null_surreason['intype'].value_counts()
```

```python
data['surreason'].value_counts()
```

```python
#fill null surreason with "N/A"
data['surreason'].fillna('LOST/FOUND', inplace=True)
```

##### Null sex

```python
data['sex'].value_counts()
```

```python
data['sex'].fillna('U', inplace=True)
```

##### Null bites

```python
data['bites'].fillna('U', inplace=True)
```

##### Null petsize

```python
null_size = null_data[null_data['petsize'].isnull()]
null_size['animaltype'].value_counts()
```

```python
null_size['breed'].value_counts()
```

```python
#check if any of the null size animals have / or mix in their breed
null_size[null_size['breed'].str.contains('/')]['breed'].value_counts()
```

```python
null_size[null_size['breed'].str.contains('mix')]['breed'].value_counts()
```

```python
#create a table of the mode petsize for each breed
mode_size = data.groupby('breed')['petsize'].agg(pd.Series.mode).to_frame()
```

```python
#using the mode table, fill the nulls with the mode for each breed
data['petsize'].fillna(data['breed'].map(mode_size['petsize']), inplace=True)
```

```python
#check for nulls in data
data.isnull().sum()
```

```python
#drop the remaining nulls
data.dropna(inplace=True)
```

#### Creating new relevant columns

```python
data
```

##### Handling mixed breeds

```python
#if the breed ends with " / MIX", contains ' / ', or contains 'MIX', then it is mixed
data['mixed'] = data['breed'].apply(lambda x: True if x.endswith(' / MIX') or ' / ' in x or 'MIX' in x else False)
```

```python
data[data['mixed']]
```

```python
#drop ' / MIX' from the end of the breed and replace with ''
data['breed'] = data['breed'].apply(lambda x: x.replace(' / MIX', ''))
```

```python
#for breeds with multiple breeds, split the breeds into two columns
data[['breed','breed2']] = data['breed'].str.split(' / ', expand=True)
data[data['mixed']]
```

```python
#if mixed is true and breed2 is 'None' replace with 'Unknown'
data.loc[(data['mixed']) & (data['breed2'].isnull()), 'breed2'] = 'Unknown'
```

```python
#fill nan in breed2 with 'N/A'
data['breed2'].fillna('N/A', inplace=True)
```

#### Creating date-connected columns

```python
#create a new column called time_in that is indate - outdate
data['days_in'] = pd.to_datetime(data['outdate']) - pd.to_datetime(data['indate'])
data['days_in'] = data['days_in'].dt.days
data
```

```python
#create two new columns in_season and out_season that are fall, winter, spring, summer based on the month of indate and outdate
data['in_season'] = pd.to_datetime(data['indate']).dt.month.map({1: 'WINTER', 2: 'WINTER', 3: 'SPRING', 4: 'SPRING', 5: 'SPRING', 6: 'SUMMER', 7: 'SUMMER', 8: 'SUMMER', 9: 'FALL', 10: 'FALL', 11: 'FALL', 12: 'WINTER'})
data['out_season'] = pd.to_datetime(data['outdate']).dt.month.map({1: 'WINTER', 2: 'WINTER', 3: 'SPRING', 4: 'SPRING', 5: 'SPRING', 6: 'SUMMER', 7: 'SUMMER', 8: 'SUMMER', 9: 'FALL', 10: 'FALL', 11: 'FALL', 12: 'WINTER'})
```

```python
#drop indate and outdate columns
data.drop(columns=['indate', 'outdate'], inplace=True)
```
